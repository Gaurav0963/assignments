```python
# ANSWER 1 : Write a Python Program to Check if a Number is Positive, Negative or Zero?

try:
    number = float(input("Enter a Real Number: "))
    if number > 0:
        print('\n{} is POSITIVE NUMBER'.format(number))
    elif number < 0:
        print('\n{} is NEGATIVE NUMBER'.format(number))
    else:
        print("\nit's ZERO!")

except ValueError:
    print('\nNumrical input needed')
```

```python
# ANSWER 2: Write a Python Program to Check if a Number is Odd or Even?

try:
    def EvenOdd(a):
        if a%2 == 0:
            print('\n{} is EVEN'.format(a))
        else:
            print('\n{} is ODD'.format(a))
    
    num = int(input("Enter Integer: "))
    EvenOdd(num)

except ValueError:
    print("\nOnly enter an Integer")
```

```python
# ANSWER 3: Write a Python Program to Check Leap Year?

def leap(year):
        
    if year%100 == 0 and year%400 == 0:
        print('\n{} is LEAP year'.format(year))
        
    elif year%100 != 0 and year%4 == 0:
        print('\n{} is LEAP year'.format(year))
        
    else:
        print('\n{} is NOT a LEAP year'.format(year))

try:    
    year_ = input('Enter year in YYYY format: ')
    ar = [int(i) for i in year_]
    yr = int(year_)
    
    if len(year_) == 4 and ar[0] != 0:
        leap(yr)
    else:
        print('\nIncorrect year Entered')
        print('1. Year should be in YYYY format')
        print('2. Year should not start with ZERO')
        
except ValueError:
    print('\nIncorrect Input!!! \nInput other than digits is not allowed')
    
```

```python
# ANSWER 4: Write a Python Program to Check Prime Number?

def prime(number) :
    check = 0
    for i in range(2, number):
        if number%i == 0:
            check += 1
            break
            
    if check > 0:
        print('\n',number,'is NOT a Prime Number')
    else:
        print('\n',number,'is a Prime Number')

try:
    num = int(input("Enter Natural Number: "))
    if num == 2:
        print('\n',num,'is the SMALLEST PRIME Number')
    elif num>1:
        prime(num)
    else:
        print('\n',num,'is NOT a Prime Number')

except ValueError:
    print('\nIncorrect Input')
        
            
```

```python
# ANSWER 5:Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

def PRIME(a, b):
    for num in range(a, b + 1):
        if num > 1:
            for i in range(2, num):
                if (num % i) == 0:
                    break
            else:
                yield (num)

result = PRIME(1, 10000)
print(*list(result), sep='\n')
```
