### Question 2: Write a Python Program to Multiply Two Matrices?

```python
import os

def matrix_input(rows, columns):
    mat = []
    for i in range(rows):
        row = []
        for j in range(columns):
            row.append(int(input('a({}, {}): '.format(i+1, j+1))))
        mat.append(row)
    return mat


def matrix_product(A, B):
    product = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    
    for a in range(len(A)):
        for b in range(len(B[0])): 
            for c in range(len(B)):            
                product[a][b] += (A[a][c]) * (B[c][b])
    
    return product
    
    

if __name__ == '__main__':

    print('\n\t\t\tMATRIX MULTIPLICATION')

    while True:
        try:
            row_1, column_1 = int(input('\nEnter Rows: ')), int(input('Enter Columns for Matrix 1: '))
            print('\nInput in Matrix 1 : Element(row, column)\n')
            matrix_1 = matrix_input(row_1, column_1)

            column_2 = int(input('Enter Columns for Matrix 2: '))
            print('\nInput in Matrix 2 : Element(row, column)\n')
            matrix_2 = matrix_input(row_1, column_2)
            
            os.system('CLS')

            print('\nMatrix 1 : ')
            for m1 in matrix_1:
                print(m1)

            print('\nMatrix 2 : ')
            for m2 in matrix_2:
                print(m2)
                
            result = matrix_product(matrix_1, matrix_2)

            print("\nThe PRODUCT of Above two Matrices is : ")
            for r in result:
                print(r)

        except ValueError as v:
            print(v)

        ch = input("\nEnter 'y' to Continue, ANY OTHER key to exit: ")
        if ch == 'y' or ch == 'Y':
            os.system('CLS')
            continue
        else:
            print('\n\t\t\tTHANKYOU')
            break
```

### Question 4: Write a Python Program to Sort Words in Alphabetic Order?

```python
def Sort_Words_Alphabetic_Order(sentence):
    words = list(i.lower() for i in sentence.split())
    words.sort()
    return words


if __name__ == '__main__':
    while True:
        try:
            string = input('\nSentence: ')
            print('\nSORTED WORDS: ', Sort_Words_Alphabetic_Order(string))
        
        except ValueError as v:
            print(v)
        
        ch = input("\nEnter 'y' to Continue, ANY OTHER key to exit: " )
        if ch == 'y' or ch =='Y':
            continue
        else:
            print('\n\t\t\tTHANKYOU')
            break

```

### Question 5: Write a Python Program to Remove Punctuation From a String?

```python
def  punctuation_remover(string):
    punctuations = '''<[{(~!@#$%^&*_-:;'".,?/\)}]>'''
    for i in string:
        if i in punctuations:
            string = string.replace(i, "")
    
    return string


if __name__ == '__main__':
    while True:
        try:
            string = input('\nSentence: ')
            print('\nAfter removing punctuations:', punctuation_remover(string))
            
        except ValueError as v:
            print(v)
        
        ch = input("\nEnter 'y' to Continue, ANY OTHER key to exit: " )
        if ch == 'y' or ch =='Y':
            continue
        else:
            print('\n\t\t\tTHANKYOU')
            break
```
